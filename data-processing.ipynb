```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
import pickle
import numpy as np
import os

def load_batch(file):
    with open(file, 'rb') as f:
        batch = pickle.load(f, encoding='bytes')
        data = batch[b'data']          
        labels = batch[b'labels']     
    return data, labels

def load_cifar10(path):
    X_train_list = []
    y_train_list = []

    for i in range(1, 6):
        file = os.path.join(path, f"data_batch_{i}")
        data, labels = load_batch(file)
        X_train_list.append(data)
        y_train_list.extend(labels)

    X_train = np.concatenate(X_train_list)
    y_train = np.array(y_train_list, dtype=np.int32)

    X_test, y_test = load_batch(os.path.join(path, "test_batch"))
    X_test = X_test.astype(np.float32)
    y_test = np.array(y_test, dtype=np.int32)

    X_train = X_train.astype(np.float32) / 255.0
    X_test = X_test.astype(np.float32) / 255.0

    X_train = X_train.reshape(-1, 3, 32, 32)
    X_test = X_test.reshape(-1, 3, 32, 32)

    return X_train, y_train, X_test, y_test


if __name__ == "__main__":
    path = "./cifar-10-python/cifar-10-batches-py"
    X_train, y_train, X_test, y_test = load_cifar10(path)

    print("Train:", X_train.shape, y_train.shape)
    print("Test:", X_test.shape, y_test.shape)
```
